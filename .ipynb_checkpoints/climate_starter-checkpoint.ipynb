# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Climate Analysis

```python
#finding last date in measurement
lastdate = session.query(func.max(Measurement.date)).first()
print(lastdate)
```

```python
#converting date to integer & breaking out by year, month, and day
lastdate2 = str(lastdate)
year = int(lastdate2[2] + lastdate2[3] + lastdate2[4] + lastdate2[5])
print (year)
```

```python
month = int(lastdate2[7] + lastdate2[8])
print(month)
```

```python
day = int(lastdate2[10] + lastdate2[11])
print(day)
```

```python
print(f'The date is (m/dd/yyyy): {month}, {day}, {year}')
```

```python
#calculate date 1 year ago from last data point in the database
#using timedelta to calculate one year prior
querydate = dt.date(year, month, day) - dt.timedelta(days=365)
print("Query Date: ", querydate )
```

```python
#design query to retrieve the last 12 months of precipitation data
#and plot the results
#perform a query to retrieve the data and precipitation scores

#convert last date to datetime.date format
maxdate = dt.date(year, month, day)

prcp_list = []
prcp_list = session.query(Measurement.date, Measurement.prcp).\
    filter(Measurement.date > querydate).filter(Measurement.date <= maxdate).\
    all()

print(prcp_list)
```

```python
#save query results as a pandas dataframe
#set index to date column
#convert list to dataframe
#convert date column data to datetime using pandas

prcpdf = pd.DataFrame(prcp_list)
prcpdf['date'] = pd.to_datetime(prcpdf['date'])
prcpdf.head()
```

```python
#set date column as index and rename precipitation column for graphing
#sort dataframe by date 
prcpdf2 = prcpdf.set_index('date')
prcpdf2.rename(columns={'prcp':'precipitation'}, inplace=True)
prcpdf2.head()
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
prcpdf2.plot(figsize=(15,8), sort_columns=True, rot=50, use_index=True, legend=True)
plt.xlabel('Date')
plt.ylabel('Precipitation')
plt.title('Precipitation Measurements for August 24, 2016 to August 23, 2017')
plt.savefig('barplotprecip')
plt.show()
```

```python
# Design a query to show how many stations are available in this dataset?
stationnum = []
stationnum = session.query(Station.station).count()
print(stationnum)
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
activestation = []
activestation = session.query(Measurement.station, Station.name,
                             func.count(Measurement.station)).\
filter(Measurement.station == Station.station).group_by(Measurement.station).order_by(func.count(Measurement.station).desc()).all()
activestation
```

```python
mostactive = activestation[0][1]
mostactivenum = activestation[0][2]
print(f'The most active station is {mostactive} with {mostactivenum} records.')
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature most active station?
activestats = []
activestats = session.query(func.min(Measurement.tobs), func.max(Measurement.tobs), func.avg(Measurement.tobs)).\
filter(Measurement.station == Station.station).filter(Station.name == mostactive).all()
print(activestats)

lowtemp = activestats[0][0]
hightemp = activestats[0][1]
avgtemp = activestats[0][2]
print(f'For {mostactive}, the lowest temperature is {lowtemp} and the highest temperature is {hightemp}. The average temperature is {avgtemp}.')
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
activestation[0][0]
```

```python
obsrv_list = []
obsrv_list = session.query(Measurement.station, Measurement.date, Measurement.tobs).\
    filter(Measurement.station == activestation[0][0]).filter(Measurement.date > querydate).\
    filter(Measurement.date <= maxdate).\
    all()

obsrv_df = pd.DataFrame(obsrv_list)
obsrv_df.head()
```

```python
#converting date to datetime
obsrv_df['date'] = pd.to_datetime(obsrv_df['date'])
obsrv_df.head()
```

```python
#plot the results from Query of the last 12 months of tobs data for this stations a histogram
obsrv_temps = obsrv_df['tobs']
plt.hist(obsrv_temps, bins=12, color='red')
plt.title('Temperature Observations for Most Active Station from August 24, 2016 to August 13, 2017', fontsize=20)
plt.ylabel('Frequency', fontsize=16)
labels = ['tobs']
plt.legend(labels)
plt.savefig('histogram1')
plt.show()
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```
